There are two surprising outliers in the Top and Bottom GDP graphs. Macao in CHina has shockingly high GDP Per Capita and CAR had a spike in 2009 followed by a rapid decline to the bottom in 2014.

Macao is a gaming district of China with a population of only 650,000. The nature of its existence, gambling/banks are a lucrative endeavor where their customers do not live in the country. Due to the high revenues and low residency its PER CAPITA GDP is much higher than most countries with populations in the multiple millions. 

CAR experienced a humanitarian crisis in 2009 through a genocidal civil war. Human Rights abuses were committed by both militant rebel groups and palace guards. The state of the war put immense pressure on their GDP.
