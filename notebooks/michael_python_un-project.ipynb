```python
internet_df2 = internet_df.drop(columns= ['Value Footnotes', 'Unnamed: 4', 'Unnamed: 5'])
internet_df2.tail(200)
```

```python
gdp_df2 = gdp_df2.rename(columns = {'Country or Area': 'Country', 'Value': 'GDP_Per_Capita'})
gdp_df2.columns
```

```python
internet_df2.columns
```

```python
internet_df2 = internet_df2.rename(columns = {'Country or Area': 'Country','Value': 'Internet_Users_Pct'})
internet_df2.columns
```

```python
type(gdp_df2)
type(internet_df2)
#checking to make sure my tables are df's
```

```python
#attempting dataframe merge below
```

```python
gdp_and_internet_use = gdp_df2.merge(internet_df2, how ='outer',on =['Country','Year'])
gdp_and_internet_use
```

```python
gdp_and_internet_use_2004_2009_2014 = gdp_and_internet_use[gdp_and_internet_use['Year'].isin([2004,2009,2014])]
gdp_and_internet_use_2004_2009_2014.head(20)
```

```python
gdp_internet_2004 = gdp_and_internet_use[gdp_and_internet_use.Year==2004]
gdp_internet_2004
```

```python
gdp_internet_2009 = gdp_and_internet_use[gdp_and_internet_use.Year==2009]
gdp_internet_2009
```

```python
gdp_internet_2014 = gdp_and_internet_use[gdp_and_internet_use.Year==2014]
gdp_internet_2014
```

```python
Top_Internet_Use_2014 = gdp_internet_2014.sort_values(by = 'Internet_Users_Pct', ascending = False).head(10)
Top_Internet_Use_2014
#Iceland had the highest percentage of internet users in 2014
```

```python
Low_Internet_Use_2014 = gdp_and_internet_use_2014.sort_values(by = 'Internet_Users_Pct', ascending = True).head(10)
Low_Internet_Use_2014
# Dem. People's Rep. of Korea (North Korea) shows 0% of users. Eritrea is right above them at .99%.
# I hesitate to count N. Korea as it appears there is no data for either category and I know it's a country with
# high censorship and a digital veil blocking outside connection
```

```python
Top_Internet_Use_2004 = gdp_and_internet_use_2004.sort_values(by = 'Internet_Users_Pct', ascending = False).head(10)
Top_Internet_Use_2004
# Sweden had the highest percent of internet users at 83.89%
```

```python
Low_Internet_Use_2004 = gdp_and_internet_use_2004.sort_values(by = 'Internet_Users_Pct', ascending = True).head(10)
Low_Internet_Use_2004
# N. Korea has the lowest percent at 0 and Myanmar just above it at .024%
```

```python
Top_Internet_Use_2009 = gdp_and_internet_use_2009.sort_values(by = 'Internet_Users_Pct', ascending = False).head(10)
Top_Internet_Use_2009
# Iceland has the top 
```

```python
Low_Internet_Use_2009 = gdp_and_internet_use_2009.sort_values(by = 'Internet_Users_Pct', ascending = True).head(10)
Low_Internet_Use_2009
# N. Korea had the lowest internet usage at 0% with Timor-Leste just above it at .18%
```

```python
High_gdp_2014 = gdp_and_internet_use_2014.sort_values(by = 'GDP_Per_Capita', ascending = False).head(10)
High_gdp_2014
# Macao had the highest GDP Per Capita in 2014 at 156514.133850
```

```python
plt.scatter(gdp_and_internet_use_2004.Internet_Users_Pct,gdp_and_internet_use_2004.GDP_Per_Capita, 
            label='2004 Internet vs GDP',
            marker='o')
plt.legend()
plt.style.use('ggplot')
plt.xlabel('Internet Users')
plt.ylabel('GDP Per Capita')
```

```python
plt.scatter(gdp_and_internet_use_2009.Internet_Users_Pct,gdp_and_internet_use_2009.GDP_Per_Capita, 
            label='2009 Internet vs GDP',
            marker='o',
            color='blue')
plt.legend()
plt.style.use('ggplot')
plt.xlabel('Internet Users')
plt.ylabel('GDP Per Capita')
```

```python
plt.scatter(gdp_and_internet_use_2014.Internet_Users_Pct,gdp_and_internet_use_2014.GDP_Per_Capita, 
            label='2014 Internet vs GDP',
            marker='o',
            color='orange')
plt.legend()
plt.style.use('ggplot')
plt.xlabel('Internet Users')
plt.ylabel('GDP Per Capita')
```

There is certainly a correlation between number of internet users and GDP. In 2004 fewer Internet Users was necessary to hit higher GDP than in 2009 and especially 2014. I suspect that the number of internet users that were using the internet for non-income based reasons (facebook, social media) increased exponentially without producing as many ways to monetize equal to the increase in users.
